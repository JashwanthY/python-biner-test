```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from lifelines import KaplanMeierFitter
import matplotlib.gridspec as gridspec
import numpy as np

# Paths
path = {
    'adam': 'data_set',
}

# Read datasets
adsl = pd.read_sas(os.path.join(path['adam'], 'adsl.xpt'), format='xport')
adtte = pd.read_sas(os.path.join(path['adam'], 'adtte.xpt'), format='xport')

# Decode byte strings to normal strings
adsl = adsl.applymap(lambda x: x.decode() if isinstance(x, bytes) else x)
adtte = adtte.applymap(lambda x: x.decode() if isinstance(x, bytes) else x)

# Filter and merge datasets
anl = adsl[(adsl['SAFFL'] == 'Y') & (adsl['STUDYID'] == 'CDISCPILOT01')]
anl = anl[['STUDYID', 'USUBJID', 'TRT01A']].merge(
    adtte[(adtte['PARAMCD'] == 'TTDE') & (adtte['STUDYID'] == 'CDISCPILOT01')],
    on=['STUDYID', 'USUBJID']
)
anl['TRT01A'] = pd.Categorical(anl['TRT01A'], categories=['Placebo', 'Xanomeline Low Dose', 'Xanomeline High Dose'])

# Convert CNSR column to event indicator (1 - CNSR in Python to match R)
anl['CNSR'] = 1 - anl['CNSR']

# Debug print to check data
print("Merged dataset head:")
print(anl.head())

# Estimate survival
kmf = KaplanMeierFitter()

# Adjusted figure size and layout
fig = plt.figure(figsize=(6, 6))  # Adjust the figure size as needed
gs = gridspec.GridSpec(3, 1, height_ratios=[0.1, 4, 1], hspace=0.2)

# Title and subtitle at the top left
fig.text(-0.15, 0.91, 'Time to Dermatologic Event by Treatment Group', ha='left', va='top', fontsize=10)

ax1 = fig.add_subplot(gs[1])
ax2 = fig.add_subplot(gs[2])

colors = ['red', 'green', 'blue']

# Plotting KM curves
for i, (name, grouped_df) in enumerate(anl.groupby('TRT01A')):
    kmf.fit(grouped_df['AVAL'], event_observed=grouped_df['CNSR'], label=name)
    kmf.plot_survival_function(ax=ax1, ci_show=True, color=colors[i], show_censors=True, censor_styles={'ms': 6, 'marker': '+'})

ax1.set_title('KM plot for Time to First Dermatologic Event: Safety population', fontweight='bold', fontsize=10)
ax1.set_xlabel('Time to First Dermatologic Event (Days)')
ax1.set_ylabel('Probability of event')
ax1.axhline(y=0.5, linestyle='--', color='gray')

# Set x-axis and y-axis ticks
ax1.set_xticks(np.arange(0, 220, 20))
ax1.set_yticks(np.arange(0, 1.1, 0.1))
ax1.set_xticks(np.arange(0, 210, 10), minor=True)
ax1.set_yticks(np.arange(0, 1.1, 0.05), minor=True)

# Add grid lines
ax1.grid(True, which='both', linestyle='-', linewidth=0.3, color='lightgray')
ax1.grid(True, which='major', linestyle='-', linewidth=0.5, color='lightgray')

# Adding legend to the right of the plot with adjusted bbox_to_anchor
legend = ax1.legend(title='TRT01A', loc='center left', bbox_to_anchor=(1.05, 0.5))

# Constructing the "At risk" table
times = np.arange(0, int(anl['AVAL'].max()) + 20, 20)
risk_table_data = {name: [] for name in anl['TRT01A'].cat.categories}

# Correctly calculate the number of individuals at risk at each time point
for name, grouped_df in anl.groupby('TRT01A'):
    kmf.fit(grouped_df['AVAL'], event_observed=grouped_df['CNSR'])
    at_risk_counts = []
    for time in times:
        count = (grouped_df['AVAL'] >= time).sum()
        at_risk_counts.append(count)
    risk_table_data[name] = at_risk_counts

# Debug print to check risk table data
print("Risk table data:")
for name, counts in risk_table_data.items():
    print(f"{name}: {counts}")

ax2.axis('off')
ax2.set_xlim(ax1.get_xlim())

# Format the "At risk" table
labels_ypos = 0.45  # Start higher to avoid overlapping with the lower part of the KM plot
label_spacing = 0.2  # Increase spacing if needed

ax2.text(-0.001, labels_ypos, 'At risk', fontweight='bold', color='black', transform=ax2.transAxes)
labels_ypos -= label_spacing

for label, counts in risk_table_data.items():
    ax2.text(-0.35, labels_ypos, f'{label:<25}', color='black', fontsize=10, transform=ax2.transAxes)
    for i, count in enumerate(counts):
        ax2.text(times[i], labels_ypos, f"{count:>3}", ha='center', color='black', fontsize=9, transform=ax2.transData)
    labels_ypos -= label_spacing

# Must save the plot using the custom save function
plt.show()  # This will display the plot directly in the Jupyter notebook
```
